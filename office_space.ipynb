```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
 
X = []
y = []
 
F, N = list(map(int, raw_input().split()))
for i in range(N):
    row = list(map(float, raw_input().split()))
    X.append(row[:F])
    y.append(row[F])
 
poly = PolynomialFeatures(degree=3)
XP = poly.fit_transform(X)
 
reg = LinearRegression()
model = reg.fit(XP, y)
 
tests = []
for i in range(int(raw_input())):
    tests.append(list(map(float, raw_input().split())))
 
TP = poly.fit_transform(tests)
print('\n'.join(map('{0:.2f}'.format, model.predict(TP))))
```
